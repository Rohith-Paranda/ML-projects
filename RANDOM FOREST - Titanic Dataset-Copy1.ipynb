```python
df1.isnull().sum()
```

```python
dummies = pd.get_dummies(df1).astype(int)
```

```python
dummies.head(10)
```

```python
dummies1 = pd.get_dummies(dummies.Survived).astype(int)
```

```python
dummies1.head(10)
```

```python
dummies1.rename(columns ={0: 'Dead', 1: 'Alive'}, inplace=True)
```

```python
dummies2 = pd.get_dummies(dummies.Pclass).astype(int)
dummies2.head()
```

```python
dummies2.rename(columns ={1:'FirstClass', 2:'SecondClass', 3:'ThirdClass'}, inplace=True)
```

```python
dummies2.head()
```

```python
df2 = pd.concat([dummies, dummies1, dummies2], axis="columns")
```

```python
df2.head()
```

```python
df2.drop(["Survived","Pclass"], axis="columns", inplace=True)
```

```python
df2.head()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
```

```python
df2.head()
```

```python
X = df2.drop(["Alive","Embarked_Q", "Sex_female"], axis="columns")
X.head()
```

```python
y = df2.Alive
y.head()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=133)
```

```python
rf = RandomForestClassifier()
```

```python
rf.fit(X_train, y_train)
ypred = rf.predict(X_test)
```

```python
acc = accuracy_score(y_test,ypred)
print(acc)
```
